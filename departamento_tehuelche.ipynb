```python
# Filtra los datos para el departamento 'Tehuelches'
filtered_data = data[data['name'] == 'Tehuelches']

# Define los rangos de edades y columnas relevantes
age_ranges = ['0-4_m', '5-9_m', '10-14_m', '15-19_m', '20-24_m', '25-29_m', '30-34_m', '35-39_m', '40-44_m',
              '45-49_m', '50-54_m', '55-59_m', '60-64_m', '65-69_m', '70-74_m', '75-79_m', '80-84_m', '85-89_m', '>95_m']
active_age_ranges_men = ['15-19_m', '20-24_m', '25-29_m', '30-34_m', '35-39_m', '40-44_m', '45-49_m',
                         '50-54_m', '55-59_m']
active_age_ranges_women = ['15-19_f', '20-24_f', '25-29_f', '30-34_f', '35-39_f', '40-44_f', '45-49_f',
                           '50-54_f', '55-59_f']
passive_age_ranges_men = ['0-4_m', '5-9_m', '60-64_m', '65-69_m', '70-74_m', '75-79_m', '80-84_m', '85-89_m', '>95_m']
passive_age_ranges_women = ['0-4_f', '5-9_f', '60-64_f', '65-69_f', '70-74_f', '75-79_f', '80-84_f', '85-89_f', '>95_f']

# Calcula los totales para hombres y mujeres en los rangos de edades específicos
active_men = filtered_data[active_age_ranges_men].sum().sum()
passive_men = filtered_data[passive_age_ranges_men].sum().sum()
active_women = filtered_data[active_age_ranges_women].sum().sum()
passive_women = filtered_data[passive_age_ranges_women].sum().sum()

# Crear el gráfico de barras apilado con los totales en las pilas
plt.figure(figsize=(10, 6))

labels = ['Pasivo', 'Activo', 'Pasivo']
men_values = [passive_men, active_men, passive_men]
women_values = [passive_women, active_women, passive_women]
total_values = [sum(values) for values in zip(men_values, women_values)]

x = range(len(labels))

plt.bar(x, men_values, label='Hombres', color='blue')
plt.bar(x, women_values, label='Mujeres', color='green', bottom=men_values)
plt.bar(x, total_values, color='gray', alpha=0.3, label='Total')

plt.xlabel('Estado')
plt.ylabel('Población')
plt.title('Distribución de Población por Estado (Pasivo - Activo - Pasivo) y Género - Departamento Tehuelches, Chubut (2010)')
plt.xticks(x, labels)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos de los censos y variaciones
years = [1947, 1960, 1970, 1980, 1991, 2001, 2010]
poblacion = [4269, 4884, 5154, 4728, 4801, 5159, 5384]
variacion = [0.00, 14.41, 5.53, -8.27, 1.54, 7.46, 4.36]

# Crear figura y ejes
fig, ax1 = plt.subplots(figsize=(10, 4))

# Gráfico de barras de la evolución poblacional
ax1.bar(years, poblacion, color='blue', label='Población')
ax1.set_ylabel('Población', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticks(years)

# Gráfico de línea de la variación intercensal en porcentaje
ax2 = ax1.twinx()
ax2.plot(years, variacion, marker='o', color='green', label='Variación (%)')
ax2.set_ylabel('Variación (%)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Etiquetas y título
plt.title('Evolución Poblacional y Variación Intercensal en el Departamento Tehuelches')
plt.xlabel('Año del Censo')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
